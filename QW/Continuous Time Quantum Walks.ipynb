```python
from pyzx import *
from fractions import Fraction
```

```python
class CTQW:
    def __init__(self):
        self.qubit_amount = 3
        self.circ = Circuit(qubit_amount=self.qubit_amount)
    def QFT(self):
        self.circ.add_gate("HAD",0)
        self.circ.add_gate("CP",0,1, phase=Fraction(1,2))
        self.circ.add_gate("HAD",1)
        self.circ.add_gate("CP",0,2, phase=Fraction(1,4))
        self.circ.add_gate("CP",1,2, phase=Fraction(1,2))
        self.circ.add_gate("HAD",2)
                           
    def IQFT(self):
        self.circ.add_gate("HAD",2)
        self.circ.add_gate("CP",1,2, phase=Fraction(-1,2))
        self.circ.add_gate("CP",0,2, phase=Fraction(-1,4))
        self.circ.add_gate("HAD",1)
        self.circ.add_gate("CP",0,1, phase=Fraction(-1,2))
        self.circ.add_gate("HAD",0)
    def UniOp(self):
        self.circ.add_gate("CNOT",1,0)
        self.circ.add_gate("CNOT",2,0)
        self.circ.add_gate("ZPhase",0,phase=Fraction(-1,16))
        self.circ.add_gate("CNOT",1,0)
        self.circ.add_gate("ZPhase",0,phase=Fraction(1,8))
        self.circ.add_gate("CNOT",2,0)
        self.circ.add_gate("CNOT",2,1)
        self.circ.add_gate("ZPhase",1,phase=Fraction(1,4))
        self.circ.add_gate("CNOT",2,1)
        self.circ.add_gate("ZPhase",2,phase=Fraction(1,8))
```

```python
ctqw = CTQW()
ctqw.QFT()
qft = ctqw.circ.to_graph()
#settings.tikzit_location = "/usr/bin/tikzit"
#tikz.tikzit(qft)
ctqw.UniOp()
ctqw.IQFT()
```

```python
g = ctqw.circ
draw(g.to_graph())
settings.tikzit_location = "/usr/bin/tikzit"
tikz.tikzit(g)

print(extract_simple(g.to_graph()).stats())
```

```python
fr = g.to_graph()
full_reduce(fr, quiet=False)
fr.normalize()
draw(fr)
settings.tikzit_location = "/usr/bin/tikzit"
tikz.tikzit(fr)
```

```python
fr_opt = extract_circuit(fr.copy())
print(fr_opt.stats(),"\nEQUAL: " , g.verify_equality(fr_opt))
draw(fr_opt)
settings.tikzit_location = "/usr/bin/tikzit"
tikz.tikzit(fr_opt)
```

```python
fr_graph = fr_opt.to_graph()
# Simple simplifications
spider_simp(fr_graph,quiet=False)
id_simp(fr_graph,quiet=False)
to_rg(fr_graph)
#The simps above dont change the graph too much so we can use the extract_simple
fr_opt = extract_simple(fr_graph)
print()
print(fr_opt.stats(),"\nEQUAL: " , g.verify_equality(fr_opt))
draw(fr_opt)
```
